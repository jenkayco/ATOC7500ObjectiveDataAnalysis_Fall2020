```python
## Plot windows in time space
T = 3
omega = np.arange(-2.*np.pi*3., 2.*np.pi*3.+.1, .1)
t = np.arange(0.01,T+0.01,0.01)

fig = plt.figure(figsize=(12,4))
plt.xlim(-0.05,T+0.05)
plt.ylim(-0.02,1.02)
b = np.ones(np.shape(t))
plt.plot(t,b,'-b', linewidth = 3, label = 'Boxcar window');
plt.plot((0,0),(0,1),'-b',linewidth = 3);
plt.plot((T,T),(0,1),'-b',linewidth = 3);
w = 0.5*(1-np.cos(2.*np.pi*t/T))
plt.plot(t,w,'-r',linewidth = 3, label = 'Hanning window');
plt.xlabel('time', fontsize = 12)
plt.ylabel('value of data (as a fraction)', fontsize = 12)
plt.title('Windows in time space', fontsize = 12)
plt.legend(frameon = False, fontsize = 12);
```

```python
### Plot response functions in frequency space
B = np.sinc(omega*T/(2.*np.pi))
W = np.sinc(omega*T/(2.*np.pi)) + (1./2.)*(np.sinc(omega*T/(2.*np.pi) + 1.) + np.sinc(omega*T/(2.*np.pi) - 1.))
Wa = np.sinc(omega*T/(2.*np.pi))
Wb = (1./2.)*(np.sinc(omega*T/(2.*np.pi) + 1.) + np.sinc(omega*T/(2.*np.pi) - 1.))
fig = plt.figure(figsize=(12,4))
plt.plot(omega,B/np.max(B),'-b',linewidth = 3, label = 'Boxcar response');
plt.xlabel('radial frequency', fontsize = 12)
plt.ylabel('spectral power', fontsize = 12)
plt.xlim(-16,16)
plt.plot(omega, W,'-r', linewidth = 3, label = 'Hanning response');
plt.plot(omega, Wb,'--r', linewidth = 3, label = 'Terms 2 & 3 of Hanning response');
plt.legend(frameon = False, loc = 'upper left', fontsize = 12);
```
